## 4.Split dataset into train-test sets with a fixed random_state.


```python
from sklearn.model_selection import train_test_split
```

```python
train_set,test_set = train_test_split(df,test_size=0.2,random_state=42)
```

```python
feature = train_set.drop('y',axis=1)
```

```python
labels = train_set['y']
```

## 3.Encode all categorical features using OneHotEncoding or LabelEncoding.


```python
from sklearn.preprocessing import OneHotEncoder

# 1. Initialize with handle_unknown='ignore'
# sparse_output=False makes it easier to work with as a standard array
onehot = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# 2. Fit and transform the training features
onehot_data = onehot.fit_transform(feature)

# 3. Prepare test features
feature_tst = test_set.drop('y', axis=1)

# 4. Transform test features (this will now skip the '94' error)
feature_tst_encoded = onehot.transform(feature_tst)
```

```python
onehot_data
```

## 5.Train Decision Tree Classifier and limit max_depth to avoid overfitting.


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
des = DecisionTreeClassifier()
```

```python
des.fit(onehot_data,labels)
```

```python
feature_tst
```


## 6.Predict on test data and generate classification report.

```python
#prediction
prediction = des.predict(feature_tst_encoded)
```

```python
from sklearn.metrics import accuracy_score, classification_report
label_tst = test_set['y']
# 1. Calculate the accuracy percentage
accuracy = accuracy_score(label_tst, prediction)

print(f"Accuracy Score: {accuracy * 100:.2f}%")
# 2. Detailed breakdown (Precision, Recall, F1-Score)
# Essential for this dataset because of the high 'no' subscription imbalance
print(classification_report(label_tst, prediction))
 
```
